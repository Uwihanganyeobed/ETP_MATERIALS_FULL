```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Sample data: house sizes (in square feet) and prices (in thousands)
X = np.array([[1500], [2000], [2500], [3000], [3500]])  # Feature: size
y = np.array([300, 400, 500, 600, 700])  # Target: price

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Make a prediction
predicted_price = model.predict([[2800]])
print(f"Predicted price for a house of 2800 sq ft: ${predicted_price[0] * 1000:.2f}")

def plot_data_and_model(X, y, model):
    """Plot the original data and the linear regression model."""
    plt.figure(figsize=(10, 6))
    
    # Scatter plot of the original data
    plt.scatter(X, y, color='blue', label='Data Points', s=100)
    
    # Line representing the linear regression model
    X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    y_pred = model.predict(X_range)
    plt.plot(X_range, y_pred, color='red', linewidth=2, label='Regression Line')
    
    # Labels and title
    plt.title('House Price Prediction')
    plt.xlabel('Size (sq ft)')
    plt.ylabel('Price (in thousands)')
    plt.legend()
    plt.grid()
    
    # Show plot
    plt.show()

# Call the plotting function
plot_data_and_model(X, y, model)
```
